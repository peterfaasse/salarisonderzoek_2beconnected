```python
import pandas as pd
import numpy as np
```

```python
#load data
df = pd.read_excel('Salarisonderzoek Financieel  Hypotheek Professionals 2021 Clean.xlsx')
```

```python
#clean data
#merge answer options
df['functie'] = df[[x for x in df.columns.to_list() if 'functie:' in x][0]]
for col in [x for x in df.columns.to_list() if 'functie:' in x][1:]:
    df['functie'] = df['functie'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'functie:' in x],  axis= 1, inplace = True)

df['pensioensvergoeding'] = df['Ja ik krijg een pensioenvergoeding'].fillna(df['Nee ik krijg geen pensioenvergoeding'])
df.drop(labels = [x for x in df.columns.to_list() if ' pensioenvergoeding ' in x],  axis= 1, inplace = True)

df['mobiliteitsbudget'] = df[[x for x in df.columns.to_list() if 'mobiliteitsbudget:' in x][0]]
for col in [x for x in df.columns.to_list() if 'mobiliteitsbudget:' in x][1:]:
    df['mobiliteitsbudget'] = df['mobiliteitsbudget'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'mobiliteitsbudget:' in x],  axis= 1, inplace = True)

df['factor_overstap'] = df[[x for x in df.columns.to_list() if 'factor_overstap:' in x][0]]
for col in [x for x in df.columns.to_list() if 'factor_overstap:' in x][1:]:
    df['factor_overstap'] = df['factor_overstap'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'factor_overstap:' in x],  axis= 1, inplace = True)

df['mediakanaal'] = df[[x for x in df.columns.to_list() if 'mediakanaal:' in x][0]]
for col in [x for x in df.columns.to_list() if 'mediakanaal:' in x][1:]:
    df['mediakanaal'] = df['mediakanaal'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'mediakanaal:' in x],  axis= 1, inplace = True)

df['soort werkgever'] = df[[x for x in df.columns.to_list() if 'soort werkgever' in x][0]]
for col in [x for x in df.columns.to_list() if 'soort werkgever' in x][1:]:
    df['soort werkgever'] = df['soort werkgever'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'soort werkgever' in x],  axis= 1, inplace = True)

#rename columns
df.rename(columns = {'Hoeveel uur bent u gemiddeld per week werkzaam?':'uur/pw',
                     'Hoeveel bent u werkzaam in uw huidige functie?':'jaar_werkzaam',
                    'Welk bruto maandsalaris ontvangt u op (basis van 40 uur) ?':'bruto_maandsalaris'},inplace = True)

#create categorical columns
df['uur_cat'] = pd.cut(df['uur/pw'], bins=[0, 24, 32, 40, 48, 56,100], include_lowest=True,labels=['20-24', '24-32', '32-40', '40-48', '49-56', '56-64'])

df['jaarwerkzaam_cat'] = pd.cut(df['jaar_werkzaam'] , 
                                bins=[0, 2, 4, 6, 8, 100], 
                                include_lowest=True,
                                labels=['0-2', '2-4', '4-6', '6-8', '8+'])

df['sal_tov_markt_cat'] = pd.cut(df['Hoe schat u de hoogte van uw salaris in, in vergelijking met de markt?'] , 
                                bins=[0, 20, 40, 66, 80, 100], 
                                include_lowest=True,
                                labels=['Ver onder markt', 'Enigszins onder markt', 'Marktconform', 'Enigszins boven markt', 'Ver boven markt'])

#create perc overstap en euro overstap salaris
df['perc_overstap'] = df['Ja vanaf zoveel ..... euro bruto per maand'].astype(float) / df['bruto_maandsalaris'].astype(float) 
df['perc_overstap'] = df['perc_overstap'].apply(lambda x: x+1 if x < 1 else x)
df['perc_overstap'] = (df['perc_overstap'] - 1) *100

df['extra_euro_overstap'] = (df['perc_overstap']/100+1) * df['bruto_maandsalaris'] - df['bruto_maandsalaris']
df['extra_euro_overstap_cat'] = pd.cut(df['extra_euro_overstap'], bins = [0,200,400,600,800,1000,999999], labels = ['0-199', ' 200-399','400-599','600-799','800-999','1000+'])

#Merge ranking reden ondernemer variabelen
df['flexibiliteit'] = df[[x for x in df.columns.to_list() if 'Flexibiliteit' in x][0]]
for col in [x for x in df.columns.to_list() if 'Flexibiliteit' in x][1:]:
    df['flexibiliteit'] = df['flexibiliteit'].fillna(df[col])

df['bedrijfsvoering '] = df[[x for x in df.columns.to_list() if 'bedrijfsvoering ' in x][0]]
for col in [x for x in df.columns.to_list() if 'bedrijfsvoering ' in x][1:]:
    df['bedrijfsvoering '] = df['bedrijfsvoering '].fillna(df[col])
    
df['Meer geld te verdienen'] = df[[x for x in df.columns.to_list() if 'Meer geld te verdienen' in x][0]]
for col in [x for x in df.columns.to_list() if 'Meer geld te verdienen' in x][1:]:
    df['Meer geld te verdienen'] = df['Meer geld te verdienen'].fillna(df[col])
    
df['ontplooien '] = df[[x for x in df.columns.to_list() if 'ontplooien ' in x][0]]
for col in [x for x in df.columns.to_list() if 'ontplooien ' in x][1:]:
    df['ontplooien '] = df['ontplooien '].fillna(df[col])
    
df['Eigen pensioen opbouwen'] = df[[x for x in df.columns.to_list() if 'Eigen pensioen opbouwen' in x][0]]
for col in [x for x in df.columns.to_list() if 'Eigen pensioen opbouwen' in x][1:]:
    df['Eigen pensioen opbouwen'] = df['Eigen pensioen opbouwen'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'ondernemer:' in x],  axis= 1, inplace = True)
```

```python
df.to_excel('FinSec 2021 Clean.xlsx')
```
