```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Salarisonderzoek Vastgoed  Makelaardij 2021 Clean.xlsx')
```

```python
df['functie'] = df[[x for x in df.columns.to_list() if 'functie:' in x][0]]
for col in [x for x in df.columns.to_list() if 'functie:' in x][1:]:
    df['functie'] = df['functie'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'functie:' in x],  axis= 1, inplace = True)
```

```python
df['pensioensvergoeding'] = df['pensioenvergoeding: Ja ik krijg een pensioenvergoeding'].fillna(df['pensioenvergoeding: Nee ik krijg geen pensioenvergoeding'])
df.drop(labels = [x for x in df.columns.to_list() if 'pensioenvergoeding: ' in x],  axis= 1, inplace = True)
```

```python
df['mobiliteitsbudget'] = df[[x for x in df.columns.to_list() if 'mobiliteitsbudget:' in x][0]]
for col in [x for x in df.columns.to_list() if 'mobiliteitsbudget:' in x][1:]:
    df['mobiliteitsbudget'] = df['mobiliteitsbudget'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'mobiliteitsbudget:' in x],  axis= 1, inplace = True)
```

```python
df['factor_overstap'] = df[[x for x in df.columns.to_list() if 'factor overstap:' in x][0]]
for col in [x for x in df.columns.to_list() if 'factor overstap:' in x][1:]:
    df['factor_overstap'] = df['factor_overstap'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'factor overstap:' in x],  axis= 1, inplace = True)
```

```python
df['mediakanaal'] = df[[x for x in df.columns.to_list() if 'mediakanaal:' in x][0]]
for col in [x for x in df.columns.to_list() if 'mediakanaal:' in x][1:]:
    df['mediakanaal'] = df['mediakanaal'].fillna(df[col])
    
df.drop(labels = [x for x in df.columns.to_list() if 'mediakanaal:' in x],  axis= 1, inplace = True)
```

```python
df.to_excel('Makelaardij 2021 Clean.xlsx')
```
